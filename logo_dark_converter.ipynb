```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
try:
    logo = np.random.rand(10, 10, 3)

    plt.figure(figsize=(14, 5))
    
    plt.subplot(1, 2, 1)
    plt.imshow(logo)
    plt.title("Simulated Logo")
    plt.axis('off')

    dark_logo = 1 - logo
    plt.subplot(1, 2, 2)
    plt.imshow(dark_logo)
    plt.title("Simulated Dark Logo")
    plt.axis('off')
    
    plt.show()
    
except FileNotFoundError:
    print("logo not found")
```

```python
try:
    logo = np.load('numpy-logo.npy')
    plt.figure(figsize=(14, 5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy Logo")
    plt.axis('off') 
    
    dark_logo = 1 - logo
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy Dark Logo")
    plt.axis('off') 
    np.save('dark_logo.npy', dark_logo)
    
except FileNotFoundError:
    print("logo file not found")
```
